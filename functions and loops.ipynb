```python
#🧠 Simple Trick to Remember Lambda

#👉 “Lambda = Shortcut function in ONE LINE”

Think like this:

# Normal function → long method
#Lambda → quick shortcut

#🔁 Compare Normal vs Lambda
       #✅ Normal Function
def add(a, b):
    return a + b


#⚡ Lambda Shortcut
add = lambda a, b: a + b

👉 Same work, less code!

🧩 Easy Formula to Build Lambda

👉 Just follow this pattern:

#    lambda arguments : expression

✔ Example:

lambda x: x*2
lambda a, b: a+b
lambda x: x**2
🚫 Common Mistakes (IMPORTANT)

❌ Wrong:

lambda a,b:
    a+b

❌ Wrong:

lambda a,b: print(a+b)   # Not useful (returns None)

✔ Correct:

lambda a,b: a+b

    
     #💡 When to Use Lambda?

Use lambda when:

Small calculation
One-time use

    
Inside functions like map(), filter()
🔥 Example (Very Useful)
nums = [1,2,3,4]

result = list(map(lambda x: x*2, nums))
print(result)

✔ Output:

[2, 4, 6, 8]
               #🧠 Final Shortcut Rule

👉 “If it fits in one line → use lambda”
👉 “If logic is big → use def”
```

```python
🧠 What is map()?

👉 map() is used to apply a function to every element in a list (or iterable).

👉 Instead of writing loops, you can do it in one line.
```

```python
#🧩 Syntax
map(function, iterable)
#function → what you want to apply
#iterable → list/tuple etc.
```

```python
#        Example 1: Using normal function
def square(x):
    return x*x

nums = [1, 2, 3, 4]

result = map(square, nums)
print(list(result))
```

```python
#   Example 2: Using lambda (shortcut)
nums = [1, 2, 3, 4]

result = map(lambda x: x*x, nums)
print(list(result))
```

```python
#    Example 2: Using lambda (shortcut)
nums = [1, 2, 3, 4]

result = map(lambda x: x*x, nums)
print(list(result))
```

```python
#  Example 3: Add 2 numbers from 2 lists
a = [1, 2, 3]
b = [4, 5, 6]

result = map(lambda x, y: x+y, a, b)
print(list(result))
```

```python
⚠️ Important Point

👉 map() does NOT directly give list
👉 It gives a map object, so always use:

list(map(...))
```

```python
    # Without map (for understanding)
nums = [1,2,3,4]
result = []

for i in nums:
      result.append(i*i)

print(result)

#👉 Same output, but more code 😅
```

```python
🧠 Shortcut Trick

👉 map = “apply function to all items”

💡 When to Use?

#                        Use map() when:

. You want to transform a list
. Same operation on all elements
. With lambda for short code
```

```python
                                               #🧠 What is zip()?

👉 zip() is used to combine multiple lists element-wise

👉 It pairs items from each list index by index
```

```python
#  🧩 Syntax
zip(list1, list2, ...)
```

```python
#           Example 1: Basic zip
a = [1, 2, 3]
b = ['a', 'b', 'c']

result = zip(a, b)
print(list(result))
```

```python
#           Example 2: Different length lists
a = [1, 2, 3]
b = ['a', 'b']

print(list(zip(a, b)))
```

```python
#Example 3: Loop with zip
names = ['Naresh', 'Ravi', 'Sita']
marks = [85, 90, 95]

for n, m in zip(names, marks):
    print(n, m)
```

```python
#     Convert to dictionary
keys = ['name', 'age']
values = ['Naresh', 22]

d = dict(zip(keys, values))
print(d)
```

```python
#    Reverse zip (Unzip)
pairs = [(1, 'a'), (2, 'b')]

a, b = zip(*pairs)

print(a)  # (1, 2)
print(b)  # ('a', 'b')
```

```python
🧠 Shortcut Trick

👉 zip = combine lists like a zipper

[1,2,3]
[a,b,c]
   ↓
[(1,a), (2,b), (3,c)]


⚠️ Important Points
Returns zip object, so use list() if needed
Stops at shortest iterable
Very useful in loops & dictionaries
```

```python
🚀 When to Use?

Use zip() when:

Combining multiple lists
Parallel iteration
Creating dictionary
```

```python
                        # 🧠 What is filter()?

👉 filter() is used to select elements from a list based on a condition

👉 It keeps only the values that return True
```

```python
                🧩 Syntax

filter(function, iterable)

. function → condition (must return True/False)
. iterable → list, tuple, etc.
```

```python
#  Example 1: Using normal function
def is_even(x):
    return x % 2 == 0

nums = [1, 2, 3, 4, 5, 6]

result = filter(is_even, nums)
print(list(result))
```

```python
# Example 2: Using lambda (shortcut)

nums = [1, 2, 3, 4, 5, 6]

result = filter(lambda x: x % 2 == 0, nums)
print(list(result))
```

```python
#Example 3: Filter positive numbers

nums = [-2, -1, 0, 1, 2, 3]

result = filter(lambda x: x > 0, nums)
print(list(result))
```

```python
⚠️ Important Point

👉 filter() returns a filter object, not a list
👉 So always use:

list(filter(...))
🧠 Shortcut Trick

👉 filter = “keep only True values”
```

```python
⚡ map vs filter (quick difference)
map() → transforms data
filter() → selects data

🚀 When to Use?

Use filter() when:

.You want to remove unwanted values
. Apply condition on list
. Clean data
```

```python
#       🧠 What is reduce()?

👉 reduce() is used to reduce a list into a single value

👉 It keeps applying a function and gives one final result

#       📦 Import Required

   #reduce() is not built-in directly, you must import it:

from functools import reduce


# 🧩 Syntax

reduce(function, iterable)


.Takes 2 values at a time
.Applies function
.Keeps reducing step by step
```

```python
# Example 1: Sum of list
from functools import reduce

nums = [1, 2, 3, 4]

result = reduce(lambda x, y: x + y, nums)
print(result)
```

```python
How it works internally
[1,2,3,4]

Step1: 1+2 = 3  
Step2: 3+3 = 6  
Step3: 6+4 = 10 
```

```python
# Example 2: Find maximum
from functools import reduce

nums = [10, 5, 20, 8]

result = reduce(lambda x, y: x if x > y else y, nums)
print(result)
```

```python
⚠️ Important Points
Returns single value
Works step-by-step
Needs functools module

🧠 Shortcut Trick

👉 reduce = “combine all → one result”

⚡ map vs filter vs reduce
Function	Work
map()	     Modify all elements
filter()	  Select some elements
reduce()	 Combine into one value  


🚀 When to Use?

Use reduce() when:

Sum, product, max, min
Aggregation operations
One final output needed


💡 Pro Tip (Important)

👉 In real coding:

Use sum(), max() instead of reduce when possible
But in interviews → reduce is asked frequently
```
